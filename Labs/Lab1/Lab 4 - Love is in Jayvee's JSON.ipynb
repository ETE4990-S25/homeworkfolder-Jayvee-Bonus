```python
import json

def pick_movies(movies, min_rating, max_year):
    matching_movies = []
    for movie in movies:
            rating = float(movie['rating_score'])
            year = int(movie['year_of_release'])
            if rating >= min_rating and year <= max_year:
                matching_movies.append(movie)
            else:
                continue
    return matching_movies

with open('romancemovies.json', 'r') as file:
    romancemovies = json.load(file)
        
min_rating = float(input("Enter the minimum rating (0-10): "))
max_year = int(input("Enter the latest release year: "))

selected_movies = pick_movies(romancemovies, min_rating, max_year)

print(f"Total movies processed: {len(romancemovies)}")
print(f"Movies matching criteria: {len(selected_movies)}")

if not selected_movies:
    print("No movies found matching the criteria.")
else:
    for movie in selected_movies:
        print(f"Selected Movie: {movie['title']}")
        print(f"Year: {movie['year_of_release']}")
        print(f"Rating: {movie['rating_score']}")
        print(f"Box Office: ${movie['box_office_amount']} million")
        print()
```
