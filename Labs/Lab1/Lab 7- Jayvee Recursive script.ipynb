```python
import os
import hashlib

# Function to compute MD5 checksum of a file
def get_checksum(file_path, hash_algorithm='md5'):
    hash_func = hashlib.new(hash_algorithm)  # MD5 or SHA-256
    try:
        with open(file_path, 'rb') as f:
            while chunk := f.read(8192):
                hash_func.update(chunk)
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")
        return None
    return hash_func.hexdigest()

# Function to search for files and find duplicates based on checksums
def find_duplicates(directory):
    file_dict = {}  # Dictionary to store file checksums and their paths
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            file_checksum = get_checksum(file_path)

            if file_checksum:
                if file_checksum not in file_dict:
                    file_dict[file_checksum] = []
                file_dict[file_checksum].append(file_path)
    
    # Display duplicates
    display_duplicates(file_dict)

# Function to display files that are identical (same checksum)
def display_duplicates(file_dict):
    found_duplicates = False
    for checksum, file_paths in file_dict.items():
        if len(file_paths) > 1:  # If there are multiple files with the same checksum
            found_duplicates = True
            print(f"\nIdentical files (checksum: {checksum}):")
            for file_path in file_paths:
                print(f"  - {file_path}")
    
    if not found_duplicates:
        print("No duplicate files found.")

# Menu function to allow the user to choose actions
def menu():
    while True:
        print("\n--- File Duplicate Finder ---")
        print("1. Enter directory to search for duplicates")
        print("2. Exit")
        choice = input("Choose an option: ").strip()

        if choice == '1':
            start_dir = input("Enter the directory to search: ").strip()
            if not os.path.isdir(start_dir):
                print("Invalid directory. Please try again.")
                continue
            find_duplicates(start_dir)
        elif choice == '2':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the menu if the script is executed directly
if __name__ == "__main__":
    menu()


```
